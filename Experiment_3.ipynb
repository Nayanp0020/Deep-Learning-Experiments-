```python
import numpy as np
import matplotlib.pyplot as plt
from google.colab import files
from tensorflow.keras.preprocessing import image
from tensorflow.keras.applications.resnet50 import ResNet50, preprocess_input, decode_predictions
```

```python
# 1. Load the pretrained ResNet50 model
# It will download weights (~100MB) from Google storage if not already present
```

```python
model = ResNet50(weights='imagenet')
```

```python
# 2. Upload your own image from your computer
uploaded = files.upload()
img_path = list(uploaded.keys())[0]  # Get the filename of the uploaded image
```

```python
# 3. Preprocess the data
# ResNet50 requires 224x224 input size
img = image.load_img(img_path, target_size=(224, 224))
x = image.img_to_array(img)
x = np.expand_dims(x, axis=0) # Add batch dimension (1, 224, 224, 3)
x = preprocess_input(x)       # Apply model-specific preprocessing
#(scaling/ centering)
```

```python
# 4. Make Predictions
preds = model.predict(x)
```

```python
# 5. Display Results
plt.imshow(img)
plt.axis('off')
plt.show()
```

```python
# Decode results into a list of tuples (class, description, probability)
print('Predicted:', decode_predictions(preds, top=3)[0])
```
