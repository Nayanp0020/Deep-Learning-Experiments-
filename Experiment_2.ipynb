```python
#Creating functions to compute various losses
```

```python
import tensorflow as tf
import numpy as np
```

```python
#Define Actual and Predicted Values
y_true = np.array([1, 0, 1, 1])

y_pred = np.array([0.9, 0.2, 0.8, 0.7])
```

```python
# Mean Squared Error (MSE)
def compute_mse(y_true, y_pred):
    mse = np.mean((y_true - y_pred) ** 2)
    return mse

print("MSE Loss:", compute_mse(y_true, y_pred))
```

```python
# Mean Absolute Error (MAE)
def compute_mae(y_true, y_pred):
    mae = np.mean(np.abs(y_true - y_pred))
    return mae

print("MAE Loss:", compute_mae(y_true, y_pred))
```

```python
#Binary Cross Entropy Loss
def compute_binary_crossentropy(y_true, y_pred):
    loss = tf.keras.losses.binary_crossentropy(y_true, y_pred)
    return tf.reduce_mean(loss)

print("Binary Crossentropy Loss:",
      compute_binary_crossentropy(y_true, y_pred).numpy())
```

```python
#Categorical Cross Entropy Loss
y_true_cat = [[1,0,0],
              [0,1,0],
              [0,0,1]]

y_pred_cat = [[0.8,0.1,0.1],
              [0.2,0.7,0.1],
              [0.1,0.2,0.7]]
```

```python
#Compute Categorical Cross Entropy
loss = tf.keras.losses.categorical_crossentropy(
    y_true_cat,
    y_pred_cat
)

print("Categorical Crossentropy Loss:",
      tf.reduce_mean(loss).numpy())
```
